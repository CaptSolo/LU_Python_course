## 1. uzdevums - klašu definēšana

- definējiet klasi Book, kam ir šādi atribūti: title, author, year
- definējiet šai klasei metodi `__init__()`, kura ļauj piešķirt norādītās vērtības šīs klases atribūtiem
- definējiet metodi `print_info()`, kas nodrukā informāciju ar šī objekta atribūtiem (piem., "Title: 1984, author: George Orwell, year: 1949")
- izveidojiet divus dažādus Book klases objektus un ar metodes print_info() palīdzību nodrukājiet informāciju par tiem
- definējiet metodi `__str__()`, kas atgriež informāciju par šo objektu
  - apskatiet kuros gadījumos šī metode nostrādā un atgriež prasīto informāciju


## 2. uzdevums - kafijas automāts

Izmantojot objekt-orientēto pieeju realizēt vienkāršotu kafijas automāta programmatūru.

- Kafijas automāts lietotājam piedāvā izvēlēties: melnu vai baltu kafiju vai tēju. Neatkarīgi no veida dzēriens maksā 80 centus.
- Lietotājs drīkst izmantot tikai 10, 20, 50 centu monētas. Automāts izdod atlikumu 10, 20 un 50 centu monētās.
- Melna kafija tiek pagatavota: 300 ml ūdens, 8 grami kafijas. Balta kafija tiek pagatavota 260 ml ūdens, 8 grami kafijas, 40 ml piena. Tēja tiek pagatavota: 300 ml ūdens, 1 tējas vienība.
- Pēc automāta apkopes tajā ir ievietoti: 5 l ūdens, 1 l piena, 5 tējas vienības, 80 g kafijas. 5 50 centu monētas, 5 20 centu monētas, 5 10 centu monētas.

Izpildīt šādus uzdevumus:

2.1) Bezmaksas automāts

1. Lietotājs izvēlas dzērienu
2. Automāts pārbauda, vai pietiek sastāvdaļu dzēriena pagatavošanai. Ja jā, tad ejam pie nākamā soļa; ja nē, tad paziņojam, ka nepietiek sastāvdaļu un process beidzas.
3. Atņemt no sastāvdaļām dzēriena pagatavošanai nepieciešamo un nodrukāt ziņu, ka kafija ir gatava.

Papildus funkcionalitāte:

- automāts atceras savu stāvokli (izejmateriālu un naudas daudzumus) starp tā palaišanas reizēm

2.2) Maksas automāts

Papildināt iepriekš izveidoto programmas kodu ar maksas automāta funkcionalitāti.

1. Lietotājs norāda, cik katra veida monētu ir ievietotas, piemēram, 8 10 centu monētas
2. Automāts pārbauda, vai pietiek naudas dzēriena iegādei. Ja jā, tad ejam pie nākamā soļa, ja nē, tad process beidzas
3. Ja naudas pietiek, tad pagatavo dzērienu, aprēķina vai nepieciešams izdot atlikumu un izdrukā atlikuma apjomu pa monētu veidiem
4. Pieskaitīt lietotāja iemaksāto naudu automātā esošajiem resursiem un atņemt atlikumā izdoto.



## 3. uzdevums - datu vizualizācija

No Latvijas atvērto datu portāla lejupielādēt meteoroloģisko novērojumu arhīva datus:
- https://data.gov.lv/dati/lv/dataset/hidrometeorologiskie-noverojumi/resource/ecc62e27-2071-483c-bca9-5e53d979faa8

Atlasīt informāciju par 2 dažādu novērojumu vietu (meteoroloģisko staciju) vidējās gaisa temperatūras rādījumiem un nodrukāt datu kopas pirmos un pēdējos ierakstus.

Vizualizēt šo novērojumu vietu vidējās gaisa temperatūras datus. Šim nolūkam var izmantot `matplotlib` vai kādu citu datu vizualizēšanas bibliotēku.
- datus var vizualizēt kādi tie ir (pa stundām) vai arī vispirms aprēķināt katras dienas vidējos rādījumus

## 4. uzdevums - attēlu datu apstrāde

Uzinstalējiet bibliotēkas darbam ar attēlu datiem (vai pārliecinieties, ka tās jau ir instalētas):
- `imageio`
- `scikit-image` vai `pillow`

Izvēlaties jebkādu krāsu attēlu pēc savas izvēles un veiciet ar to šādas darbības:

1. Ar `imageio.v2.imread` palīdzību nolasiet šo attēla failu
2. Aplūkojiet šo attēlu ar matplotlib.pyplot `imshow` funkciju
3. Uzlabojiet attēla kontrastu (scikit-image `exposure.equalize_hist`)
4. Izpludiniet šo attēlu (scikit-image `filters.gaussian`)
5. Uzlabojiet attēla asumu (scikit-image `unsharp_mask`)
